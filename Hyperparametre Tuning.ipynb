```python
# test train split

y = df['booking_status']

x = df.drop('booking_status',axis=1)

x_train, x_test, y_train, y_test = train_test_split(x, y, test_size=0.30, random_state=0)
```

```python
rndf = RandomForestClassifier(n_estimators=100, min_samples_leaf =5,max_leaf_nodes =5, max_depth=5, random_state=0)
rndf.fit(x_train, y_train)
y_pred=rndf.predict(x_test)

print('accuracy_score', accuracy_score(y_test, y_pred))
print('f1_score', f1_score(y_test, y_pred))
```

```python
r = export_text(rndf.estimators_[0], feature_names=list(df.iloc[:,:5].columns))
print(r)
```

```python
rndf = RandomForestClassifier(n_estimators=100, max_depth=10,  random_state=0)
scores = cross_val_score(rndf, x, y, cv=5)
print(scores)

print((scores.mean(), scores.std()))
```

```python
grid_param = {
    'n_estimators': [90,100],
    'criterion': ['gini'],
    'max_depth': range(2,7,1),
    'min_samples_leaf': range(1,7,1),
    'min_samples_split': range(1,7,1),
    'max_features': ['log2']

}
```

```python
grid_search = GridSearchCV(estimator=rndf, param_grid=grid_param, cv=5, n_jobs=-1,verbose=3)

grid_search.fit(x_train, y_train)
```

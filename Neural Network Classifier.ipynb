```python
def min_max_scaling(data, feature_range=(0, 1)):
    min_val, max_val = feature_range
    return (data - np.min(data)) / (np.max(data) - np.min(data)) * (max_val - min_val) + min_val
```

```python
df=min_max_scaling(df, feature_range=(0,1))
```

```python
y = df['Diabetes']
x = df.drop(columns='Diabetes')
x_train, x_test, y_train, y_test = train_test_split(x, y, test_size=0.30, random_state=0)
```

```python
clf = MLPClassifier(solver='lbfgs', alpha=1e-5, hidden_layer_sizes=(5, 2), random_state=1)
clf.fit(x_train, y_train)
clf.predict(x_test)

print('accuracy_score', accuracy_score(y_test, clf.predict(x_test)))
print('recall', recall_score(y_test, clf.predict(x_test)))
print('f1_score', f1_score(y_test, clf.predict(x_test)))
```

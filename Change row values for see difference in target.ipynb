```python
dtc = DecisionTreeClassifier(max_depth=3, random_state=0)
dtc.fit(x_train, y_train)
y_pred=dtc.predict(x_test)

print('accuracy_score', accuracy_score(y_test, y_pred))
print('f1_score', f1_score(y_test, y_pred))
```
